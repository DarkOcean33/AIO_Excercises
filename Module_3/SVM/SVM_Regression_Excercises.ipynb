```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    mean_absolute_error, 
    mean_squared_error,
)
```

```python
dataset_path = './Content/auto_insurance.csv'
df = pd.read_csv(
    dataset_path,
    names=[
        'n_claims',
        'total_payment'
    ]
)
```

```python
normalizer = StandardScaler()
df_normalized = normalizer.fit_transform(df)
```

```python
df_normalized
```

```python
X, y = df_normalized[:, 0], df_normalized[:, 1]
X = X.reshape(-1, 1)
```

```python
val_size = 0.2
text_size = 0.125
random_state = 0
is_shuffle = True

X_train, X_val, y_train, y_val = train_test_split(
    X,
    y,
    test_size=val_size,
    random_state=random_state,
    shuffle=is_shuffle
)

X_train, X_test, y_train, y_test = train_test_split(
    X_train,
    y_train,
    test_size=text_size,
    random_state=random_state,
    shuffle=is_shuffle
)
```

```python
print(f'Number of training samples: {X_train.shape[0]}')
print(f'Number of validation samples: {X_val.shape[0]}')
print(f'Number of test samples: {X_test.shape[0]}')
```

```python
regressor = SVR()
regressor.fit(X_train, y_train)
```

```python
y_val_pred = regressor.predict(X_val)
y_test_pred = regressor.predict(X_test)
val_mae = mean_absolute_error(y_val, y_val_pred)
val_mse = mean_squared_error(y_val, y_val_pred)
test_mae = mean_absolute_error(y_test, y_test_pred)
test_mse = mean_squared_error(y_test, y_test_pred)

print('Evaluation results on validation and test set:')
print(f'Mean Absolute Error\tVal: {val_mae}\tTest: {test_mae}')
print(f'Mean Squared Error\tVal: {val_mse}\tTest: {test_mse}')
```
