```python
import numpy as np
from numpy import genfromtxt
import matplotlib.pyplot as plt
data = np.genfromtxt('./Content/advertising.csv', delimiter =',', skip_header =1)
N = data.shape[0]
print(N)
X = data[:, :3]
y = data[:,3:]
     
```

```python
def mean_normalization(X):
  N = len(X)
  maxi = np.max(X)
  mini = np.min(X)
  avg = np.mean(X)
  X_b = (X - avg) / (maxi - mini)
  X_b = np.hstack([np.ones((N, 1)), X_b])
  return X_b, maxi, mini, avg

X_b, maxi, mini, avg = mean_normalization(X)
```

```python
def stochastic_gradient_descent(X_b, y, epochs =50, learning_rate=0.00001):
  thetas = np.array([[1.16270837], [-0.81960489], [1.39501033], [0.29763545]])
  thetas_path = [thetas]
  losses = []


  for epoch in range(epochs):
    for i in range(N):
      random_index = i
      xi = np.array(X_b[random_index:random_index+1])
      yi = y[random_index: random_index+1]

      y_hat = xi.dot(thetas)
      y_hat = y_hat

      loss = (y_hat - yi) * (y_hat - yi)/2
      loss = loss.ravel()
      losses.append(loss)

      gradient_loss = (y_hat - yi)

      compute_gradient = xi.T.dot(gradient_loss)

      thetas = thetas - learning_rate * compute_gradient
      thetas_path.append(thetas)
  return thetas_path, losses
```

```python
sgd_theta, losses = stochastic_gradient_descent(X_b, y, epochs =50, learning_rate =0.01)
```
